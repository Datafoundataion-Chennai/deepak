```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = np.arange(1,11)
stock_price = [150, 152, 149, 155, 160, 162, 158, 165, 170, 175]
trading_volume = [1.2, 1.4, 1.1, 1.8, 2.0, 2.5, 2.2, 2.7, 3.0, 3.5]
fig, ax = plt.subplots()
ax2 = ax.twinx()
ax.plot(days,stock_price,marker='o', linestyle="--", color="blue", linewidth=2, label="Stock price (USD)")
ax.set_xlabel('Days')
ax.set_ylabel('Stock Price')
ax2.plot(days,trading_volume,marker='o', linestyle="--", color="red", linewidth=2, label="Trading Volume(Millions)")
ax2.set_ylabel('Trading Volume')
plt.legend(['stock_price','trading_volume'])
plt.tight_layout()
plt.show()
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
rent = np.array([1200, 1200, 1200, 1200, 1250, 1250])
food = np.array([400, 420, 450, 480, 500, 520])
entertainment = np.array([150, 200, 220, 180, 250, 300])
transport = np.array([100, 120, 130, 110, 140, 160])
plt.bar(months,rent,color='yellow')
plt.bar(months,food,bottom=rent,color='blue')
plt.bar(months,transport,bottom=rent+food,color='red')
plt.bar(months,entertainment,bottom=rent+food+transport,color='violet')
plt.xlabel('Months')
plt.ylabel('Monthly expenses')
plt.legend(['rent','food','entertainment','transport'])
plt.show()
```

```python
weeks = np.arange(1, 9)  
covid_cases = [500, 700, 1200, 2000, 3000, 4500, 6000, 8000]  
vaccination_rate = [5, 10, 20, 35, 50, 65, 75, 85]

fig, ax = plt.subplots(2, 1)

ax[0].plot(weeks,covid_cases, marker= 'o',color="red", linestyle="--", label="Covid Cases")
ax[0].legend()
ax[0].set_title('Covid Cases')
ax[0].grid(True)

ax[1].bar(weeks,vaccination_rate, color="blue")
ax[1].legend('vaccination_rate')
ax[1].set_title('Vaccination Rate')

plt.tight_layout()
plt.show()
```
